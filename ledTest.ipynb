```python
from led import Led

cubes = [Led()]

cubeStr = []
cubeStr.append("F731, FEC8, 137F, 8CEF")

cubes[0].cubify(cubeStr[0])

for c in range(12):
    cubes.append(cubes[c].cp())# [1].rotateCW()
    cubes[c+1].rotateCCW()
    cubes[c+1].ledScatter()


# Printing Cubes
for i, c in enumerate(cubes):
    print('\nCube: ', i)
    c.printCube()

# Printing ScatterPlot data
print("\nScatter Plot: \n")
for i, c in enumerate(cubes):
    print("Cube ", i, "Scatter Length: ", len(c.LEDplot[0]))
    for s in c.LEDplot:
        print(s)
        print()

```
